```python
#H8 Classify Numbers into Even and Odd Lists
def sort_even_odd(input_list):
    even_list = []
    odd_list = []

    for num in input_list:
        if num % 2 == 0:
            even_list.append(num)
        else:
            odd_list.append(num)

    even_list.sort()
    odd_list.sort()

    return even_list, odd_list

# Example usage
input_numbers = [9, 2, 7, 4, 5, 8, 3, 6, 1]
sorted_even, sorted_odd = sort_even_odd(input_numbers)

print("Original list:", input_numbers)
print("Sorted even numbers:", sorted_even)
print("Sorted odd numbers:", sorted_odd)
```

```python
#Write a program to find the key in a dictionary that has the maximum value. Input: {‘a’: 10, ‘b’: 20, ‘c’: 15} Output: ‘b’
d = {'a':10, 'b':20, 'c': 15}
res = max(d, key=d.get)
print(res)
```
